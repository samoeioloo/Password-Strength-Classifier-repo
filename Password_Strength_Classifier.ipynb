```python
#Imbalanced data. 1 has highest count, 0and2 has lowest count
# ML model will be biased towards 1
#password:independent, strength: dependent. Separate this two

password_tuple=np.array(data) #puts data into an array.Allows retrieval of data in forms of indep/dep
```

```python
password_tuple
```

```python
import random
random.shuffle(password_tuple) #increases robustness
```

```python
#access independent
x=[labels[0] for labels in password_tuple]
#access dependent
y=[labels[1] for labels in password_tuple]
```

```python
x #all passwords
```

```python
y #all strengths
```

```python
#in order to perform TFIDF, data needs to be in character form
#function splits input into characters.

def wordDivideChar(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
wordDivideChar('visi7k1yr') #test the function on sample password
```

```python
#pass data into TFvectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=wordDivideChar) #param specify on what basis you are tokenizing the string
```

```python
#use vectorizer to achieve matrix
X=vectorizer.fit_transform(x)
```

```python
#check for shape of matrix ie (rows,cols)
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
#obtain importance of features
firstDocumentVector=X[0] #object of the vector
```

```python
#fetch its data in form of columns
firstDocumentVector.T.todense()
```

```python
#prep data for modelling purpose where first col feature, 2n col importance wrt each feature
#define dataframe
df =pd.DataFrame(firstDocumentVector.T.todense(), index=vectorizer.get_feature_names(), columns=['TF-IDF'])
#params:1-data, 2=index, 3-columns
df.sort_values(by=['TF-IDF'],ascending=False) #sorts data by highest TFIDF (frequency)
```

```python
#split data for train/test
from sklearn.model_selection import train_test_split
```

```python
XTrain, XTest, YTrain, yTest=train_test_split(X,y,test_size=0.2) #20% of data is used for testing
```

```python
XTrain.shape #first val is 80% of the initial X.shape result
```

```python
#implement ML model (LogisticRegression)
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0, multi_class='multinomial') #multinomial= because more than two sub-categories (0,1,2 respectiviely)
```

```python
#fit data
clf.fit(XTrain,YTrain)
#returns default parameters selected by LogisticRegression
```

```python
#perform prediction on data

dt=np.array(['%!@123abcd'])#consider rare case on data not in dataset
#transform data so the ML model can understand it
pred=vectorizer.transform(dt)
clf.predict(pred) #prediction
```
